```python
datos.columns = datos.columns.str.lower()
```

```python
votos = datos.groupby(['año','tipele','municipio']).sum()
```

```python
votos.fillna(0,inplace=True)
```

```python
votos.columns
```

```python
v_1 = votos[['pan', 'pri', 'prd', 'pt', 'pvem',
       'mc', 'na', 'morena', 'es', 'votval', 'votnul', 'vottot']]
```

```python
v_1['mayor'] = v_1[['pan', 'pri', 'prd', 'pt', 'pvem',
       'mc', 'na', 'morena', 'es','votnul']].idxmax(axis=1).values
```

```python
import plotly
import cufflinks

v_1.reset_index()[v_1.reset_index().año >2010].groupby(
    ['año','tipele']).mayor.value_counts().unstack().iplot(kind='bar', asUrl=True)
```
